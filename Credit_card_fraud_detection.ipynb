# Dealing with imbalanced data

# under-sampling

Build a sample dataset containing similar distribution of normal and fraud transaction

Since our classes are highly skewed we should make them equivalent in order to have a normal distribution of the classes.

```python
#Lets shuffle the data before creating the subsamples
df = df.sample(frac=1)
```

```python
# amount of fraud classes 492 rows
fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492]
```

```python
normal_distributed_df = pd.concat([fraud_df, non_fraud_df])
```

```python
# Shuffle dataframe rows
new_df = normal_distributed_df.sample(frac=1, random_state=42)
```

```python
new_df.head()
```

```python

sns.countplot('Class', data=new_df, palette=colors)
plt.title('Equally Distributed Classes', fontsize=14)
plt.show()
```

```python
new_df['Class'].value_counts()
```

# Spliting the data into training data and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2,random_state=50)
```

# Model Building

## logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

```python
classifier = LogisticRegression()
```

training the logistic Regression model with training data

```python
Logistic_model = classifier.fit(X_train,Y_train)
```

```python
Logistic_model_pred = Logistic_model.predict(X_test)
```

Model Evaluation

```python
confusion_matrix(Y_test,Logistic_model_pred)
```

```python
tn,fp,fn,tp=confusion_matrix(Y_test,Logistic_model_pred).ravel()
```

```python
print(tn,fp,fn,tp)
```

```python
recall = tp/(tp+fn)
precision = tp/(tp+fp)
```

```python
f_score = 2*recall*precision/(recall+precision)
f_score
```

```python
LR_Acc=accuracy_score(Y_test,Logistic_model_pred)
LR_Acc
```

```python
roc_auc_score(Y_test,Logistic_model_pred)
```

# KNN

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
num_folds = 10
kfold = KFold(n_splits=10)
```

# Grid Search for Tuning the no. of neighbors(K-value)



```python
from sklearn.model_selection import GridSearchCV
```

```python
n_neighbors = nm.array(range(1,80))
param_grid = dict(n_neighbors = n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator= model,param_grid=param_grid)
grid.fit(X,Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
k_range = range(1,100)
k_scores = []
# use iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=5)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

The best accuracy we are getting with KNN model is 93% when k is 6
